```python
import os 
import scipy.misc
import numpy as np
import matplotlib.pyplot as plt

from keras.models import load_model
```

### Helpers

```python
def load_img(path):
    # read image
    img = scipy.misc.imread(path, mode='RGB').astype(np.float)
    #  resize
    img = scipy.misc.imresize(img, (256,256))
    return img


def process_image(img):
    img = img.astype("float32")
    img /= 255.0
    return np.expand_dims(img, axis=0)


def inference(path):
    img = load_img(path)
    img = process_image(img)
    pred = model.predict(img)
    print(pred)
    pred = np.argmax(pred)
    
    if pred == 1:
        return "Normal."
    else:
        return "COVID-19!"
```

### Get image

```python
# Load image
path = "covid.jpeg"
image = load_img(path)
plt.imshow(image)
```

#### Get and load model

```python
# Get model weights
os.system("wget https://github.com/hasibzunair/synthetic-covid-cxr-gen/releases/download/v1.0/cxr_normalvcovid+g1g2.h5.h5")

model = None
model = load_model("cxr_normalvcovid+g1g2.h5")
model.summary()
```

```python
label = inference(path)
print(label)
```
